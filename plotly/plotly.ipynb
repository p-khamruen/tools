```python
import requests
import plotly.express as px
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/p-khamruen/tools/main/plotly/women_volleyball_player.csv')
print(data.info())
```

```python
data.head()
```

```python
df = data[(~data["Ranking in 2023"].isnull()) & (data["ActivePlayer"]=='Yes')].sort_values(by=["Ranking in 2023"])
df
```

```python
df2 = df[["Name","Nationality","Position","Ranking in 2023"]]
```

```python
def get_country_code(x):
    # print(x)
    r=''
    try:
      return pycountry.countries.get(name=x).alpha_3
    except:
      c = x.split(",")
      for idx,w in enumerate(c):
        w=w.strip()
        try:
          r += pycountry.countries.get(name=w).alpha_3
        except:
          if(w=='Russia'):
            r += 'RUS'
          elif(w=='USA'):
            r += 'USA'
          elif(w=='South Korea'):
            r += 'KOR'
          elif(w=='Turkey'):
            r += 'TUR'
          elif(w=='Czech Republic'):
            r += 'CZE'
          elif(w=='Kazakstan'):
            r += 'KAZ'
          elif(w=='Iran'):
            r += 'IRN'
          elif(w=='Vietnam'):
            r += 'VNM'
          elif(w=='Venezuela'):
            r += 'VEN'
          elif(w=='Kosovo'):
            r += 'XKX'
          elif(w=='Chinese Taipei'):
            r += 'TPE'
          elif(w=='Bolivia'):
            r += 'BOL'
          elif(w=='Moldova'):
            r += 'MDA'
          elif(w=='Cape Verde'):
            r += 'CPV'
          else:
            r += 'NoVal'
        if(idx<len(c)-1):
          r+=","
      return r
```

```python
df2.loc[:, 'iso_alpha']  = df2['Nationality'].apply(get_country_code)
new_df = df2['iso_alpha'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).to_frame('iso_alpha')
final_df = df2.drop('iso_alpha', axis=1).join(new_df).reset_index(drop=True).sort_values(by=["Ranking in 2023"])
final_df = final_df.query("iso_alpha != 'NoVal'").drop_duplicates()
final_df
```

```python
grouped_df = final_df.groupby(['iso_alpha', 'Position']).size().reset_index(name='Count').sort_values(by=["iso_alpha"], ascending=False)
grouped_df
```

```python
# df = px.data.gapminder()
df = grouped_df
fig = px.scatter_geo(df, locations = "iso_alpha", color="Position",
                     hover_name="iso_alpha",size="Count",
                     projection="natural earth")
fig.show()
```
